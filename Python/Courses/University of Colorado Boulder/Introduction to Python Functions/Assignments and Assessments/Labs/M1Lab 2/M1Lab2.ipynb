###Task: define a function that print "hello world"

###Task: define a function that print integers 1 to 10 (hint: you can use a for loop)

```python
def one_to_ten():
    for n in range(1,11,1):
        print(n)

one_to_ten()
```

###Task: define a function that print primary integers in 100 (hint: you can use loops)

```python
def print_primes_up_to_100():
    for n1 in range (2, 101):
        is_prime = True
        for n2 in range(2,n1):
            if n1 % n2 == 0:
                is_prime = False
                break
        if is_prime:
            print(n1)

print_primes_up_to_100()

#thought hard here lol
```

###Task: define a function that print even number from 1 to 10.

```python
def even_numbers_1_to_10():
    for n in range(1,11):
        if n % 2 == 0: print(n)

even_numbers_1_to_10()
```

###Task: define a function that print odd number from 1 to 10

```python
def odd_numbers_1_to_10():
    for n in range(1, 11):
        if n % 2 != 0:
            print(n)


odd_numbers_1_to_10()
```
