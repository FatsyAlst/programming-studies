# Module 2: **Variables and operations**

## Introduction to variables

### Variables
- What are variables?
  - You can imagine a **variable** is a **reference** of a **data**
  - When you want to **compute** the data, you can use the **variable** instead
    - When you use **x** in a computation, Python will use the **data** that **x** is referring

### Identifiers of variables
- Variables have **identifiers** - their names
  - Each **identifier** may consist of **letters (A-Z, a-z)**, **digits (0-9)** andd **underscores (`_`)**
  - Identifiers must **begin** with a letter or the underscore
  - Identifiers may **not** begin with a digit
  - Identifiers cannot have special characters
  - Identifiers are case sensitive
  - Identifiers may not be reserved keywords

### Naming conventions
- Naming variables is not just for correct
- We follow naming conventions for **readability**
  - The name should be meaningful so readers can understand
  - You can connect multiple words use underscore, or capitalize the 1st letter from second words
    - average_income, start_location, first_name
    - totalSale, profitByQuarter, lastName

## Assignment operation
- We use `=` to assign a value to a variable
  - Variable name on the left of `=`
  - Value on the right of `=`
  - This **single** equations is called **assignment operation**
- `message = 'Hello, world!'` 
  - A value **`'Hello, world!'`** is assigned to variable **message**
  - `print(message)` will be the same as `print("Hello, world!")`

````python
message = 'Hello, world!'
print(message)

print("message")
````

# Module 3 **Flow control - Branching**

## What is flow control?

### Flow control
- **Flow control** is the **logical** design that controls **order** in which set of statements execute
- So far, the programs we wrote, are executing line by line - or **sequence structure**
- We need more:
  - **Branching** - decision structure
  - **Repetition** - looping structure

## Branching and `if` statements

### Branching
- We can have specific action(s) performed only if a **condition** is satisfied
  - Also knows as **decision** structure, or **selection** structure

### The `if` statement
- We use `if` to start a branching:
```
if condition:
    statement(s)
```
- The first line knows as the `if` clause:
  - Includes the keyword `if`
  - Specifies the **condition** with a `:`
  - List statement(s) with one **indention**
  - Only when the condition is tested to be True, the statement(s) will be executed; otherwise, the block will be skipped

````python
# compute the absolute value

x = float(input("Please enter a number: "))
print(x)

if x < 0:
    x = -x
print('Absolute value of x is:', x)
````

## Multiple branches

### Branching
- We can have alternative action(s) performed based on a **condition**

### The `f-else` statement
- We can also use `if-else` to start a branching:
```
if condition:
    statements
else:
    statements
```
- `if` clause and `else` clause must be aligned:
  - `if` clause specifies the **condition** with a `:`
  - `else` clause with a `:`
  - List statement(s) with consistent **indention**
  - If the condition is tested to be True: the statement(s) in `if` clause will be executed; otherwise, the statement(s) in `else` clause will be executed

````python
# even or odd checking

x = int(input("Please enter a positive integer: "))

if x % 2 == 0:
    print(x, 'is even')
else:
    print(x, 'is odd')
````

## Decision tree

### Branching
- We can have a specific action(s) performed based on a **nested condition**

### The `if-elif-else` statement
- We can also use `if-elif-else` so start a branching:
```
if condition1:
    statement(s)
elif condition2:
    statement(s)
else:
    statement(s)
```
- `if` clause, `elif` clause, and `else` clause must be aligned
  - `if` clause and `elif` cause specify the **conditions** with a `:`
  - `else` clause with a `:`
  - List statement(s) with consistent **indention**
  - If the condition is tested to be True, the statement(s) in *if/elif* clause will be executed; otherwise, the statement(s) in `else` clause will be executed

````python
# Hiking> if temp > 90 it is too hot, if temp < 60 it is too cold

temp = float(input("Please enter the temperature of today: "))

if temp > 80:
    print("No, it's too hot!")
elif temp < 60:
    print("No, it's too cold")
else:
    print("Perfect! Let's go!")
````

````python
# letter grade A, B, C, D, F are correspondig to [90, 100], [80, 89], [70, 79], [60, 69], [0, 59]

points = float(input("Please enter the point grade of the student [0, 100]: "))

if points >= 90:
    print('A')
elif points >= 80:
    print('B')
elif points >= 70:
    print('C')
elif points >= 60:
    print('D')
else:
    print('F')
````

# Module 4: **Flow control - Repetition**

## Why do we need repetitions?

### Repetition
- We often need to run programs (or sub-tasks) repeatedly
- Rather than write the same code again and again, we can set a **repetition** flow control to execute statements based on:
  - **Condition**
  - **Count**

### #Condition-based-repetition
- While the condition is tested to be True, the enclosed statements will be executed

### The `while` loop
- We can use `while` to start a loop:
```
while condition:
    statement(s)
```
- The first line known as the `while` clause
  - `while` clause specifies the **condition** with a `:`
  - List statement(s) with consistend **indention**
  - If the condition is tested to be True, the statement(s) in `while` clause will be executed; otherwise, skip and move to next

````python
# print all positive integers up to 10

x = 0
while x <= 10:
    print(x)
    x += 1
````

````python
# print all integers from 10 to 0

x = 10
while x >= 0:
    print(x)
    x -= 1
````

````python
# input validation - only print 'bingo' when typed 'secret'

password_on_record = 'secret'

password_entered = input("Please enter your password: ")

while password_entered != password_on_record:
    password_entered = input("Wrong password! Try again: ")

print('bingo')
````

````python
# flag - enter 'stop' to exit a program

flag = True

while flag:
    inputs  = input("Please enter some words to display, enter STOP to exit: ")
    if inputs == 'STOP':
        flag = False
    print(inputs)
````

## Count-based repetition
- We may also need to repeat the program (or sub-tasks) for a specific number of times

### The `for` loop
- We can use `for` to start a loop:
```
for item in [item1, item2, ... , itemn]:
    statement(s)
```
- The first line known as the `for` clause:
  - `for` clause specifies the **sequence** of **itens** with a `:`
  - The **number** of items **n** in the sequence is the number of time the statements will be executed
  - List statement(s) with consistent **indention**

### The `for` loop with `range`
- We can use `range` to create a sequence for a `for` loop:
```
for item in range(n):
    statement(s)
```
- `range()` characteristics:
  - `range(0,5,1)` -> [0,1,2,3,4] (start, up to but not including, step)

````python
# print a list of integers [1,2,3,5,7]

for i in [1, 2, 3, 5, 7]:
    print(i)
````

````python
# print a range of numbers

for i in range(5, 10, 3):
    print(i)
````

````python
# print each char in a string

for c in 'Hello, world!':
    print(c)
````

````python
# calculate sum of a list of numbers

numbers = [1, 2, 3, 5, 7]

sum = 0
for n in numbers:
    sum += n

print(sum)
````

````python
# calculate sum of 0 to 100

sum = 0
for n in range(101):
    sum += n

print(sum)
````

## `continue` and `break`
- In repetition structure, we can use two special control for the repetitions:
  - `continue`
    - Skip the rest of the statements and jump to next iteration
  - `break`
  - Stop the repetition immediately

````python
# when enter stop to exit

while True:
    inputs = input("Please enter some word to display and 'STOP' to stop")
    if inputs == 'STOP':
        break
    print(inputs)
````

````python
# print all numbers whose square is less than 100

x = 0
while True:
    if x ** 2 >= 100:
        break
    print(x)
    x += 1
````

````python
# print numbers dividable by 5 within 100

for i in range(101):
    if i % 5 != 0:
        continue
    print(i)
````
