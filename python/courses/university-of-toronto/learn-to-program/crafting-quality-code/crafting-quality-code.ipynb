````python
# Algorith 1: reverse the string; compare the reversed string to the original string

def reverse(word):
    """(str) -> str
    
    Return a reversed version of word.
    
    >>> reverse('hello')
    'olleh'
    >>> reverse('a')
    'a'
    """
    
    rev = ''
    
    # For each character in word, add that character to the beginning of rev
    for ch in word:
        rev = ch + rev
    
    return rev


def is_palindrome_v1(word):
    """(str) -> bool
    
    Return True if and only if word is a palindrome.
    
    >>> is_palindrome_v1('noon')
    True
    >>> is_palindrome_v1('racecar')
    True
    >>> is_palindrome_v1('dented')
    False
    """
    
    return reverse(word) == word

print(is_palindrome_v1('noon'))
print(is_palindrome_v1('racecar'))
print(is_palindrome_v1('dented'))

# We can rewrite the reverse(word) function changing the return statement to only word = word[::-1]
````

````python
# Algorithm 2: Split the sring into two halves, reverse the second half and compare the first half to the reversed second half

def reverse(word):
    """(str) -> str
    
    Return a reversed version of word.
    
    >>> reverse('hello')
    'olleh'
    >>> reverse('a')
    'a'
    """
    
    return word[::-1]

def is_palindrome_v2(word):
    """(str) -> bool
    
    Return True if and only if word is a palindrome.
    
    >>> is_palindrome_v2('noon')
    True
    >>> is_palindrome_v2('racecar')
    True
    >>> is_palindrome_v2('dented')
    False
    """
    
    n = len(word) # number of characters in word
    
    # Comparte the first half of word to the reverse of the second half.
    # Omit the middle character of an odd-length string.
    return word[:n // 2] == reverse(word[n - n // 2:])


print(is_palindrome_v2('noon'))
print(is_palindrome_v2('racecar'))
print(is_palindrome_v2('dented'))
````

````python
# Algorithm 3: Comparte the 1st character to the last character; compare the 2nd character to the second last character (crop when the middle of the string is reached)

def is_palindrome_v3(word):
    """(str) -> bool
    
    Return True if and only if word is a palindrome.
    
    >>> is_palindrome_v3('noon')
    True
    >>> is_palindrome_v3('racecar')
    True
    >>> is_palindrome_v3('dented')
    False
    """
    
    i = 0
    j = len(word) - 1
    
    while i < j and word[i] == word[j]:
        i += 1
        j -= 1
    
    return j <= i


print(is_palindrome_v3('noon'))
print(is_palindrome_v3('racecar'))
print(is_palindrome_v3('dented'))
````

````python
def recommend(file, price, cuisines_list):
    """(file open for reading, str, list of str) -> list of [int, str]
    
    Find restaurants in file that are priced according to price and that are tagged
    with any of the items in cuisines_list. Return a list of [int, str] of the
    form [rating%, restaurant name], sorted by rating%
    """
    
    # Read the file and build the data structures
    # - a dict of {restaurant name: rating%}
    # - a dict of {price: list of restaurant names}
    # - a dict of {cuisine: list of restaurant names}
    name_to_rating, price_to_names, cuisine_to_names = read_restaurants(file)
    
    # Price: look up the list of restaurant names for the requested price.
    names_matching_price = price_to_names[price]
    
    # Now we have a list of restaurants in the right price range
    # Need a new list of restaurants that serve one of the cuisines
    names_final  = filter_by_cuisine(names_matching_price, cuisine_to_names, cuisines_list)
    
    # Now we have a list of restaurants that are in the right price range and serve the requested cuisine
    # Need to look at ratings and sort this list
    result = build_rating_list(name_to_rating, names_final)
    
    # We're done! Return that sorted list
    return result


def build_rating_list(name_to_rating, names_final):
    """(dict of {str: int}, list of str) -> list of [int, str]
    
    >>> name_to_rating = {'Georgia Porgie': 87,
    'Queen St. Cafe': 82,
    'Dumplings R Us': 71,
    'Mexican Grill': 85,
    'Deep Fried Everything: 52'}
    >>> names = ['Queen St. Cafe', 'Dumplings R Us']
    >>> build_rating_list(name_to_rating, names)
    [[82, 'Queen St. Cafe'], [71, Dumplings R Us]]
    """
    
    sorted_list = []
    for restaurant in names_final:
        if restaurant in name_to_rating:
            sorted_list.append([name_to_rating[restaurant], restaurant])
    sorted_list.sort(reverse=True)
    
    return sorted_list


def filter_by_cuisine(names_matching_price, cuisine_to_names, cuisines_list):
    """(list of str, dict of {str: list of str}, list of str) -> list of str
    
    >>> names = ['Queen St. Cafe', 'Dumplings R Us', 'Deep Fried Everything']
    >>> cuis = {'Canadian': ['Georgie Porgie'],
    'Pub Food': ['Georgie Porgie', 'Deep Fried Everything'],
    'Malaysian': ['Queen St. Cafe'],
    'Thai': ['Queen St. Cafe'],
    'Chinese': ['Dumplings R Us'],
    'Mexican': ['Mexican Grill']}
    >>> cuisines = ['Chinese', 'Thai']
    >>> filter_by_cuisine(names, cuis, cuisines)
    ['Queen St. Cafe', 'Dumplings R Us']
    """
    
    restaurants_with_cuisines = []
    for cuisine in cuisines_list:
        if cuisine in cuisine_to_names:
            for restaurant in cuisine_to_names[cuisine]:
                if restaurant in names_matching_price and restaurant not in restaurants_with_cuisines:
                    restaurants_with_cuisines.append(restaurant)
    
    return restaurants_with_cuisines

def read_restaurants(file):
    """(file) -> (dict, dict, dict)
    
    Return a tuple of three dictionaries based on the information in the file
    - a dict of {restaurant name: rating%}
    - a dict of {price: list of restaurant names}
    - a dict of {cuisine: list of restaurant names}
    """
    
    name_to_rating = {}
    price_to_names = {'$': [], '$$': [], '$$$': [], '$$$$': []}
    cuisine_to_names = {}
    
    with open(file, "r") as fhand:
        lines = [line.strip() for line in fhand.readlines() if line.strip() != '']
    
    for i in range(0, len(lines), 4):
        if i + 3 < len(lines):
            
            name = lines[i]
            rating = float(lines[i + 1][:-1])
            price = lines[i + 2]
            cuisines = [c.strip() for c in lines[i + 3].split(',')]
            
            name_to_rating[name] = rating
            price_to_names[price].append(name)
            
            for cuisine in cuisines:
                if cuisine not in cuisine_to_names:
                    cuisine_to_names[cuisine] = []
                    cuisine_to_names[cuisine].append(name)
                else:
                    cuisine_to_names[cuisine].append(name)
        
    return name_to_rating, price_to_names, cuisine_to_names

print(recommend(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\restaurants.txt', '$$', ['Chinese', 'Mexican', 'Thai']))
````

You will be given just some docstrings of incomplete functions. From there on, you are on your own again!

````python
def read_motorbikes(file):
    """(file) -> (dict, dict, dict, dict, dict)
    
    Return a tuple of five dictionaries based on information in the file:
    - a dict of {motorbike name: year}
    - a dict of {power level: list of motorbike names}
    - a dict of {engine cc: list of motorbike names}
    - a dict of {bike_type: list of motorbike names}
    - a dict of {usage_category: list of motorbike names}
    - a dict of {motorbike name: price}
    
    >>> # Example structure
    >>> name_to_year = {'Honda CB 500F': 2023, 'Yamaha YZF-R1': 2023}
    >>> power_to_names = {'Low': [...], 'Medium': [...], 'High': [...]}
    """
    
    # Opening the file and storing the lines (excluding newlines) in a list for better data analysis
    with open(file, "r") as fhand:
        lines = [line.strip() for line in fhand if line.strip() != '']
    
    # Creating some variables to store the infos
    name_to_year = {}
    power_to_names = {}
    cc_to_names = {}
    type_to_names = {}
    usage_to_names = {}
    name_to_price = {}
    
    # Reading the infos for each bike and storing these variables
    for i in range(0, len(lines), 7):
        if i + 6 < len(lines):
            name = lines[i]
            year = lines[i + 1]
            power = lines[i + 2]
            cc = lines[i + 3]
            type = lines[i + 4]
            usages = lines[i + 5].split(',')
            price = float(lines[i + 6])
    
        # Storing these infos
        name_to_year[name] = year
        if power not in power_to_names:
            power_to_names[power] = []
            power_to_names[power].append(name)
            if name not in power_to_names[power]:
                power_to_names[power].append(name)
        else:
            power_to_names[power].append(name)
        if cc not in cc_to_names:
            cc_to_names[cc] = []
            cc_to_names[cc].append(name)
            if name not in cc_to_names[cc]:
                cc_to_names[cc].append(name)
        if type not in type_to_names:
            type_to_names[type] = []
            type_to_names[type].append(name)
            if name not in type_to_names[type]:
                type_to_names[type].append(name)
        for usage in usages:
            if usage not in usage_to_names:
                usage_to_names[usage] = []
                usage_to_names[usage].append(name)
            if name not in usage_to_names[usage]:
                usage_to_names[usage].append(name)
        name_to_price[name] = price

    return name_to_year, power_to_names, cc_to_names, type_to_names, usage_to_names, name_to_price


def filter_by_usage(names_matching_power, usage_to_names, usage_list):
    """(list of str, dict of {str: list of str}, list of str) -> list of str
    
    Filter motorcycles by usage categories that match the power range.
    Return motorcycles that serve at least one usage category 
    from usage_list AND are in the power range, without duplicates.
    
    >>> names = ['Honda CB 500F', 'Yamaha YZF-R1']
    >>> usage_dict = {'Racing': ['Yamaha YZF-R1'], 
    ...               'Commuting': ['Honda CB 500F', 'Kawasaki Ninja 250']}
    >>> filter_by_usage(names, usage_dict, ['Racing'])
    ['Yamaha YZF-R1']
    
    """
    
    # After easily filtering by power leve, we need to read the usage_to_names
    # dict and return other list of bikes that have the usage in usage_list and
    # store these infos in a final_names variable
    
    # Creating the variable to store the filtered bikes
    final_names = []

    # We look at the lists of bikes in usage_to_names by using the desired usages as index
    for usage in usage_list:
        # We iterate through the lists of the wanted usages and add the info (if they aren't already) in the final_list
        for bike in usage_to_names[usage]:
            if bike in names_matching_power and bike not in final_names:
                final_names.append(bike)
    return final_names


def recommend_motorbikes(file, power_level, usage_list):
    """(file open for reading, str, list of str) -> list of str, float
    
    Find motorcycles with the specified power level that are suitable for 
    the given usage categories. Return a list of motorbike name 
    sorted alphabetically by motorbike name.
    
    >>> recommend_motorbikes('motorbikes.txt', 'Medium', ['Commuting'])
    [['Honda CB 500F', 2023.0], ['Suzuki GSX-S750', 2022.5]]
    """
    
    # Read the motorbikes.txt file and store the date in some variables
    name_to_year, power_to_names, cc_to_names, type_to_names, usage_to_names, name_to_price = read_motorbikes(file)
    
    # Filter the bikes easily according to the desired power_level
    names_matching_power = power_to_names[power_level]
    
    # Filter again according to the desired usage(s)
    final_names = filter_by_usage(names_matching_power, usage_to_names, usage_list)
    
    # Return that final list (sorted by alphabetical order - we do a data analysis because it's case sensitive)
    final_names.sort(key = str.lower)
    
    return final_names


print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'Medium', ['Commuting']))
# Esperado: ['Harley-Davidson Street 750', 'Honda CB 500F', 'Suzuki GSX-S750']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'High', ['Racing']))
# Esperado: ['Yamaha YZF-R1']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'Low', ['Commuting']))
# Esperado: ['Kawasaki Ninja 250', 'KTM 390 Duke']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'High', ['Touring']))
# Esperado: ['BMW F 850 GS']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'Medium', ['Sports']))
# Esperado: ['Honda CB 500F', 'Suzuki GSX-S750']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'Low', ['Racing']))
# Esperado: []

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'Medium', ['Touring']))
# Esperado: ['Harley-Davidson Street 750']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'High', ['Off-Road']))
# Esperado: ['BMW F 850 GS']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'Medium', ['Commuting', 'Touring']))
# Esperado: ['Harley-Davidson Street 750', 'Honda CB 500F', 'Suzuki GSX-S750']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'High', ['Track-Day']))
# Esperado: ['Yamaha YZF-R1']

print(recommend_motorbikes(r'C:\Users\felip\Documents\GitHub Repositories\programming-studies\Python\Courses\University of Toronto\Learn to Program - Crafting Quality Code\Files\motorbikes.txt', 'Low', ['Commuting', 'Sports']))
# Esperado: ['Kawasaki Ninja 250', 'KTM 390 Duke']
````

## Testing automatically using `unittest`

### What is `unittest`?

- A Python module to organize and automate testing of individual pieces ("units") of your programâ€”usually functions or classes
- Similar to test systems in other languages (Java's JUnit, etc.)
- **Writing unit tests = writing code to check other code automatically**

### How does it work?

You write a test class (inherits from `unittest.TestCase`), and define methods whose names start with `test_`

**Example**:
```python
import unittest

def add(a, b):
    return a + b

class TestAddFunction(unittest.TestCase):
    def test_add_positive(self):
        self.assertEqual(add(2, 3), 5)

    def test_add_negative(self):
        self.assertEqual(add(-2, -3), -5)

    def test_add_zero(self):
        self.assertEqual(add(0, 5), 5)
```

**How do you run the tests?**
- Usually at the end of your file, you include:
```python
if __name__ == '__main__':
    unittest.main(exit=False)
```
- Then you run your file
- `unittest` finds all subclasses of `unittest.TestCase` and runs all methods named `test_*`

### What can you check?

You use **assertion methods**, like:
- `self.assertEqual(a, b)` â€” checks if `a == b`
- `self.assertTrue(x)` â€” checks if `x` is `True`
- `self.assertFalse(x)` â€” checks if `x` is `False`
- `self.assertRaises(Exception, func, *args)` â€” checks if an exception is raised

Full list: https://docs.python.org/3/library/unittest.html#assert-methods

### Why is this useful?

- **Automated**: Always checks for bugs when you re-run tests
- **Structured**: Shows which test failed and why
- **Re-usable**: Tests run every time you change your code

### Typical File Structure

```python
# mylib.py (your code file)
def multiply(x, y):
    return x * y

# test_mylib.py (test file)
import unittest
from mylib import multiply

class TestMultiply(unittest.TestCase):
    def test_positive(self):
        self.assertEqual(multiply(2, 3), 6)
    
    def test_zero(self):
        self.assertEqual(multiply(0, 2), 0)

if __name__ == '__main__':
    unittest.main(exit=False)
```

### The `exit=False` parameter

**What does `exit=False` do?**
- By default, `unittest.main()` calls `sys.exit()` at the end, which exits the Python process
- Setting `exit=False` prevents `sys.exit()` from being called, letting the Python process continue

**When is this useful?**
- **Interactive environments**: Jupyter notebooks, IPython console, VSCode Interactive window - you don't want your kernel/session to shut down!
- **Scripts with more logic**: If you have code to run after tests

```python
import unittest

class MyTests(unittest.TestCase):
    def test_example(self):
        self.assertTrue(True)

if __name__ == '__main__':
    unittest.main(exit=False)
    print("This will print after tests finish (because we used exit=False).")
```

Without `exit=False`, the print statement won't run since Python exits after the tests.

### Real-World Tips

- Put tests in a separate `tests/` folder
- Use meaningful names: `test_add_with_negative_numbers`
- Run tests before every commit for reliable code
- Many tools and editors (VSCode, PyCharm) automatically detect and run unittest files

### Why is `unittest` better than `doctest`?

**Power and flexibility**:
- `unittest` lets you write complex test logic, handle various cases, test exceptions, edge cases, error handling
- `doctest` only checks whether input/output in docstrings matches exactly - great for simple examples, less for real complexity

**Organization**:
- `unittest` supports grouping tests into classes and files, with setup/teardown code
- `doctest` jumps from docstring to docstring, hard to organize or scale

**Test coverage**:
- `unittest` can cover code that doesn't fit into examples or isn't public/API-facing
- `doctest` relies on examples in docstrings - if you forget an edge case, it won't be tested

**Output and debugging**:
- `unittest` shows which test failed, why, and can print detailed info
- `doctest` typically just says "example failed" - output format must match exactly

#### The Problem: Mutating Lists

Suppose you want to remove from one list all items that appear in another list.  
Algorithm: For every item in the second list, if the item appears in the first list, remove it.

##### Example Implementation

```python
def remove_shared(L1, L2):
    """ (list, list) -> NoneType
    Remove items from L1 that are in both L1 and L2.

    >>> list_1 = [1, 2, 3, 4, 5, 6]
    >>> list_2 = [2, 4, 5, 7]
    >>> remove_shared(list_1, list_2)
    >>> list_1
    [1, 3, 6]
    >>> list_2
    [2, 4, 5, 7]
    """
    for v in L2:
        if v in L1:
            L1.remove(v)
```

- The mutation effect must be checked by examining the mutated object afterwards.
- For thorough testing, check that only objects specified to be mutated actually change.

### What is Binary Search?

Binary search is a **much faster algorithm** for finding a value in a **sorted list**.

**Key requirement:** The list **must be sorted** from smallest to largest.

***

### How Binary Search Works

Binary search works by repeatedly **dividing the search space in half**.

#### The Core Idea

Imagine a sorted list with 100,000 items:

```
| items less than v | v | items greater than v |
```

1. Look at the **middle item** (call it `v`)
2. Compare it to the value you're searching for
3. Since the list is sorted:
   - Everything to the **left** of `v` is â‰¤ `v`
   - Everything to the **right** of `v` is â‰¥ `v`
4. If your target is **greater than** `v`, eliminate the entire left half (50,000 items!)
5. If your target is **less than** `v`, eliminate the entire right half
6. Repeat on the remaining half

**With just one comparison, we eliminate half the list!**

***

### The Binary Loop Invariant

At all times during binary search, our list is divided into **three regions**:

```
        b                    m                               e
| items less than v | unknown items | items greater than or equal to v |
```

- **`b`** (beginning): marks the start of the unknown section
- **`e`** (end): marks the end of the unknown section
- **`m`** (middle): the middle index we check each iteration

***

### Visual Step-by-Step Example

Let's search for `5` in the sorted list `[2, 3, 5, 7, 11, 13]`:

#### Initial State

```

[2, 3, 5, 7, 11, 13]
 ^                ^
 b                e
```

All items are unknown at the start.

#### Step 1: Check Middle

```
m = (0 + 5) // 2 = 2
[2, 3, 5, 7, 11, 13]
       ^
       m
```

`L[2] = 5`, which equals our target â†’ **Found!**

***

```

       m

```

`5 < 11`, so our target must be to the **right**  
â†’ Set `b = m + 1 = 3`

#### Step 2

```
[2, 3, 5, 7, 11, 13]
          ^       ^
          b       e
m = (3 + 5) // 2 = 4
[2, 3, 5, 7, 11, 13]
             ^
             m
L[4] = 11
```

**Found!** Return index `4`

### Implementation

```python
def binary_search(L, v):
    """ (list, object) -> int
    
    Precondition: L is sorted from smallest to largest, and
    all the items in L can be compared to v.
    
    Return the index of the first occurrence of v in L, or
    return -1 if v is not in L.
    
    >>> binary_search([2, 3, 5, 7], 2)
    0
    >>> binary_search([2, 3, 5, 5], 5)
    2
    >>> binary_search([2, 3, 5, 7], 8)
    -1
    """
    
    b = 0              # beginning of the unknown section
    e = len(L) - 1     # end of the unknown section
    
    while b <= e:
        m = (b + e) // 2    # middle index
        
        if L[m] < v:
            b = m + 1       # v must be in the right half
        else:
            e = m - 1       # v must be in the left half or at m
    
    # At this point, b > e, so the unknown section is empty
    if b == len(L) or L[b] != v:
        return -1
    else:
        return b
```

***

### Understanding the Final If Statement

When the `while` loop ends (`b > e`), there are **two possible reasons**:

1. **All items in the list are less than `v`**  
   â†’ `b` will be equal to `len(L)` (outside the list)  
   â†’ Return `-1`

2. **The value might be at index `b`**  
   â†’ Check if `L[b] == v`  
   â†’ If yes, return `b`  
   â†’ If no, return `-1`

***

### Why Does This Work?

The key insight is the **binary loop invariant**:

- Everything to the **left** of `b` is **less than** `v`
- Everything to the **right** of `e` is **greater than or equal to** `v`
- When `b > e`, the unknown section is empty
- If `v` is in the list, it **must be at index `b`**

4. **Preconditions matter**

````python
def binary_search(L, v):
    """ (list, object) -> int
    
    Precondition: L is sorted from smallest to largest, and
    all the items in L can be compared to v.
    
    Return the index of the first occurrence of v in L, or
    return -1 if v is not in L.
    
    >>> binary_search([2, 3, 5, 7], 2)
    0
    >>> binary_search([2, 3, 5, 5], 5)
    2
    >>> binary_search([2, 3, 5, 7], 8)
    -1
    """
    
    b = 0              # beginning of the unknown section
    e = len(L) - 1     # end of the unknown section
    
    while b <= e:
        m = (b + e) // 2    # middle index
        
        if L[m] < v:
            b = m + 1       # v must be in the right half
        else:
            e = m - 1       # v must be in the left half or at m
    
    # At this point, b > e, so the unknown section is empty
    if b == len(L) or L[b] != v:
        return -1
    else:
        return b


# Test cases
print("Binary Search Examples:")
print(binary_search([2, 3, 5, 7], 2))      # Expected: 0
print(binary_search([2, 3, 5, 5], 5))      # Expected: 2
print(binary_search([2, 3, 5, 7], 8))      # Expected: -1
print(binary_search([1, 3, 5, 7, 9, 11], 7))   # Expected: 3
print(binary_search([1, 3, 5, 7, 9, 11], 1))   # Expected: 0
print(binary_search([1, 3, 5, 7, 9, 11], 11))  # Expected: 5

# Run doctests
import doctest
doctest.testmod()
````

## [Bubble Sort Algorithm](https://d3c33hcgiwev3.cloudfront.net/_d2c907ae088ca932e542255748a566c2_bubble.html?Expires=1764710491&Signature=EhBErUvGbcED4JKmxjMqDEVYzRTm5GTNX~EVx0pw0MoZTdvRUnv7xiS1M8LCXXnwu1hqhzR8WrkzDfBd1D991LRiaYVoJzf2VgeHWBBJCgwkIQs8cdf-jzgJjYhrQTq0KDEQ5W9b8~dMqGfIufYHkxLH7ayzY~Nj0Ghq4pbFbL0_&Key-Pair-Id=APKAJLTNE6QMUY6HBC5A)

**Definition**: Bubble Sort is a simple sorting algorithm that repeatedly compares adjacent items and swaps them if they are in the wrong order. The largest item "bubbles up" to its correct position in each pass.

### How it Works

1. Start at the beginning of the list (index `i = 0`)
2. Compare adjacent items at indices `i` and `i + 1`
3. If `L[i] > L[i + 1]`, swap them
4. Move to the next pair (`i = i + 1`)
5. Repeat until reaching the end of the unsorted section
6. After each complete pass, the largest unsorted item is in its correct position
7. Continue passes until the entire list is sorted

### Visual Example

**Initial List**: `[7, 3, 5, 2]`

**Pass 1** (end = 3):
- Compare 7 and 3 â†’ swap â†’ `[3, 7, 5, 2]`
- Compare 7 and 5 â†’ swap â†’ `[3, 5, 7, 2]`
- Compare 7 and 2 â†’ swap â†’ `[3, 5, 2, 7]`
- Result: 7 is now in correct position

**Pass 2** (end = 2):
- Compare 3 and 5 â†’ no swap â†’ `[3, 5, 2, 7]`
- Compare 5 and 2 â†’ swap â†’ `[3, 2, 5, 7]`
- Result: 5 is now in correct position

**Pass 3** (end = 1):
- Compare 3 and 2 â†’ swap â†’ `[2, 3, 5, 7]`
- Result: List is sorted!

### Key Concepts

**Unsorted vs Sorted Sections**:
- The list is divided into unsorted (left) and sorted (right) sections
- Variable `end` tracks the last index of the unsorted section
- After each pass, `end` decreases by 1

**Swapping Technique**:
```python
L[i], L[i + 1] = L[i + 1], L[i]  # Python's elegant swap syntax
```

### Time Complexity

- **Worst Case**: O(nÂ²) - when list is in reverse order
- **Best Case**: O(nÂ²) - even if already sorted (basic version)
- **Average Case**: O(nÂ²)

**Why O(nÂ²)?**
- Outer loop runs n-1 times (number of passes)
- Inner loop compares adjacent items
- Total comparisons â‰ˆ n Ã— n = nÂ²

````python
def bubble_sort(L):
    """(list) -> NoneType
    
    Sort the items of L from smallest to largest.
    
    >>> L = [7, 3, 5, 2]
    >>> bubble_sort(L)
    >>> L
    [2, 3, 5, 7]
    """
    
    # The index of the last unsorted item
    end = len(L) - 1
    
    while end != 0:
        # Bubble once through the unsorted section to move the largest item to index end
        for i in range(end):
            if L[i] > L[i + 1]:
                # Swap if current item is greater than next item
                L[i], L[i + 1] = L[i + 1], L[i]
        
        # Decrease end - the largest item is now in correct position
        end = end - 1

# Test the function
if __name__ == '__main__':
    import doctest
    doctest.testmod()
    
    # Additional manual test
    test_list = [7, 3, 5, 2]
    print(f"Before sorting: {test_list}")
    bubble_sort(test_list)
    print(f"After sorting: {test_list}")
````

Let's trace the algorithm with `L = [7, 3, 5, 2]`:

**Initial State**:
- `end = 3` (last index)
- Unsorted section: indices 0-3

**Pass 1** (`end = 3`):
- `i = 0`: Compare L[0]=7 and L[1]=3 â†’ swap â†’ `[3, 7, 5, 2]`
- `i = 1`: Compare L[1]=7 and L[2]=5 â†’ swap â†’ `[3, 5, 7, 2]`
- `i = 2`: Compare L[2]=7 and L[3]=2 â†’ swap â†’ `[3, 5, 2, 7]`
- `end = 2` (7 is now sorted)

**Pass 2** (`end = 2`):
- `i = 0`: Compare L[0]=3 and L[1]=5 â†’ no swap â†’ `[3, 5, 2, 7]`
- `i = 1`: Compare L[1]=5 and L[2]=2 â†’ swap â†’ `[3, 2, 5, 7]`
- `end = 1` (5 is now sorted)

**Pass 3** (`end = 1`):
- `i = 0`: Compare L[0]=3 and L[1]=2 â†’ swap â†’ `[2, 3, 5, 7]`
- `end = 0` (3 is now sorted)

**Loop exits** (`end == 0`)

Final result: `[2, 3, 5, 7]` âœ…

### Important Notes

1. **In-place sorting**: Bubble sort modifies the original list (no return value)
2. **Stable sort**: Equal elements maintain their relative order
3. **Simple but inefficient**: Good for teaching, bad for large datasets
4. **Comparison-based**: Uses `>` operator to compare elements

## [Selection Sort](https://d3c33hcgiwev3.cloudfront.net/_7df8bd0bb2645b35a137e7df1c862577_selection.html?Expires=1764711506&Signature=CeBMwla9QYtPjz2O5NNNpThuoWCLwa2-95ytphzC2pH6qQ3O~OnAJ7EsaMJh~Hw8o4xz0zTkLNub5V0ek25HtvtnceCAqY-drEHe2A0s-yMcQAApxfhj5pXzYOgU-UIRtWki3YqqFh~U6lk7WLjCIeRMMAjzD4yn3cdiotGpBMk_&Key-Pair-Id=APKAJLTNE6QMUY6HBC5A)

### Algorithm Overview

**Selection Sort** works by repeatedly finding the smallest element in the unsorted portion of the list and moving it to the sorted portion.

#### Key Characteristics:
- **Time Complexity:** O(nÂ²) - quadratic time (same as Bubble Sort)
- **Space Complexity:** O(1) - sorts in-place
- **Stability:** Not stable (can change the relative order of equal elements)
- **Best use case:** Small lists or when memory is limited

---

### Visual Example

Starting list: `[3, 7, 2, 5]`

#### Pass 1: Find smallest in entire list
```
i
[3, 7, 2, 5]  â†’ Find smallest: 2
[2, 7, 3, 5]  â†’ Swap with position i (index 0)
```

#### Pass 2: Find smallest in unsorted part
```
    i
[2, 7, 3, 5]  â†’ Find smallest in [7, 3, 5]: 3
[2, 3, 7, 5]  â†’ Swap with position i (index 1)
```

#### Pass 3: Find smallest in remaining unsorted part
```
       i
[2, 3, 7, 5]  â†’ Find smallest in [7, 5]: 5
[2, 3, 5, 7]  â†’ Swap with position i (index 2)
```

#### Pass 4: Only one element left
```
          i
[2, 3, 5, 7]  â†’ List is sorted!
```

---

### Time Complexity Analysis

- **Outer loop:** Runs `n` times (for each position)
- **Inner loop:** Runs `n-1`, `n-2`, ..., `1` times
- **Total comparisons:** $\frac{n(n-1)}{2} = O(n^2)$
- **Same complexity as Bubble Sort**, but typically performs better because it makes fewer swaps (only one swap per pass)

````python
def get_index_of_smallest(L, i):
    """(list, int) -> int
    
    Return the index of the smallest item in L[i:].
    
    >>> get_index_of_smallest([2, 7, 3, 5], 1)
    2
    >>> get_index_of_smallest([5, 4, 3, 2, 1], 0)
    4
    >>> get_index_of_smallest([10, 20, 30], 2)
    2
    """
    # The index of the smallest item so far
    index_of_smallest = i
    
    # Check each item in the unsorted part (from i+1 to end)
    for j in range(i + 1, len(L)):
        if L[j] < L[index_of_smallest]:
            index_of_smallest = j
    
    return index_of_smallest


def selection_sort(L):
    """(list) -> NoneType
    
    Sort the items of L from smallest to largest using selection sort.
    Modifies L in-place.
    
    >>> L = [3, 7, 2, 5]
    >>> selection_sort(L)
    >>> L
    [2, 3, 5, 7]
    >>> L = [5, 4, 3, 2, 1]
    >>> selection_sort(L)
    >>> L
    [1, 2, 3, 4, 5]
    """
    # For each position in the list
    for i in range(len(L)):
        # Find the index of the smallest item in L[i:] and swap that
        # item with the item at index i
        index_of_smallest = get_index_of_smallest(L, i)
        L[index_of_smallest], L[i] = L[i], L[index_of_smallest]


if __name__ == '__main__':
    import doctest
    doctest.testmod()
    
    # Manual test
    test_list = [3, 7, 2, 5]
    print(f"Original list: {test_list}")
    selection_sort(test_list)
    print(f"Sorted list: {test_list}")
````

## [Insertion Sort](https://d3c33hcgiwev3.cloudfront.net/_614ff751b1c239e03c23a377b9ec9d20_insertion.html?Expires=1764721034&Signature=LfYNf9a4I3VYqBRbB~h72KHEC3X22SasLnf~CO7SeXKy4vFctF32Z4h73acmpnqjgig8z72xcmF~1efhd~1MF4SKBFftDs7LBeKmVRGyop0W~OMPsJVKOCFc4cBHSO1kxqPXR9BItetDxnHw-cmYxVu~v8FdKZ52fjStZL2t4nM_&Key-Pair-Id=APKAJLTNE6QMUY6HBC5A)

### Algorithm Overview

**Insertion Sort** builds the sorted list one element at a time by inserting each element into its correct position within the already-sorted portion.

#### Key Characteristics:
- **Time Complexity:** O(nÂ²) worst case, but O(n) best case for nearly sorted data
- **Space Complexity:** O(1) - sorts in-place
- **Stability:** Stable (maintains relative order of equal elements)
- **Best use case:** Small lists or nearly sorted data

---

### Visual Example

Starting list: `[3, 7, 2, 5]`

#### Pass 1: Start with first element (already "sorted")
```
 i
[3, 7, 2, 5]  â†’ First element is already sorted, move to next
```

#### Pass 2: Insert second element into sorted part
```
    i
[3, 7, 2, 5]  â†’ 7 > 3, already in correct position
```

#### Pass 3: Insert third element (2) into sorted part
```
       i
[3, 7, 2, 5]  â†’ Need to insert 2 into [3, 7]
[3, 7, _, 5]  â†’ Shift 7 to the right
[3, _, 7, 5]  â†’ Shift 3 to the right
[2, 3, 7, 5]  â†’ Insert 2 at the beginning
```

#### Pass 4: Insert fourth element (5) into sorted part
```
          i
[2, 3, 7, 5]  â†’ Need to insert 5 into [2, 3, 7]
[2, 3, 7, _]  â†’ Shift 7 to the right
[2, 3, 5, 7]  â†’ Insert 5 in correct position
```

---

### How It Works

**Key Idea:** Maintain a sorted portion at the start of the list and grow it one element at a time.

1. Start with the second element (first is already "sorted")
2. Compare it with elements in the sorted part (from right to left)
3. Shift larger elements to the right
4. Insert the element in its correct position
5. Repeat for all remaining elements

---

### Time Complexity Analysis

- **Best case:** O(n) - when list is already sorted (only one comparison per element)
- **Average case:** O(nÂ²) - random order
- **Worst case:** O(nÂ²) - when list is reverse sorted
- **Better than Bubble/Selection Sort** for nearly sorted data!

````python
def insert(L, i):
    """(list, int) -> NoneType
    
    Precondition: L[:i] is sorted from smallest to largest.
    
    Move L[i] to where it belongs in L[:i + 1].
    
    >>> L = [7, 3, 5, 2]
    >>> insert(L, 1)
    >>> L
    [3, 7, 5, 2]
    >>> L = [1, 3, 2, 4]
    >>> insert(L, 2)
    >>> L
    [1, 2, 3, 4]
    """
    # The value to be inserted into the sorted part of the list
    value = L[i]
    
    # Find the index, j, where the value belongs.
    # Make room for the value by shifting.
    j = i
    while j != 0 and L[j - 1] > value:
        # Shift L[j - 1] one position to the right to L[j]
        L[j] = L[j - 1]
        j = j - 1
    
    # Put the value where it belongs
    L[j] = value


def insertion_sort(L):
    """(list) -> NoneType
    
    Sort the items of L from smallest to largest using insertion sort.
    Modifies L in-place.
    
    >>> L = [7, 3, 5, 2]
    >>> insertion_sort(L)
    >>> L
    [2, 3, 5, 7]
    >>> L = [3, 7, 2, 5]
    >>> insertion_sort(L)
    >>> L
    [2, 3, 5, 7]
    """
    # For each position in the list
    for i in range(len(L)):
        insert(L, i)


if __name__ == '__main__':
    import doctest
    doctest.testmod()
    
    # Manual test
    test_list = [3, 7, 2, 5]
    print(f"Original list: {test_list}")
    insertion_sort(test_list)
    print(f"Sorted list: {test_list}")
````

### Important Notes

#### Helper Function: `insert()`
- Takes a list `L` and index `i`
- **Precondition:** `L[:i]` must already be sorted
- Inserts `L[i]` into its correct position in `L[:i+1]`
- Uses a **while loop** to shift elements right until finding correct position

#### The Shifting Process
```python
while j != 0 and L[j - 1] > value:
    L[j] = L[j - 1]  # Shift element right
    j = j - 1         # Move left
```
- Starts from position `i` and moves left
- Shifts larger elements to the right
- Stops when finding correct position or reaching start of list

## Sorting Algorithm Pattern Recognition

| Algorithm | Sorted Portion Grows | What Happens Each Pass | Key Pattern to Spot |
|-----------|---------------------|------------------------|---------------------|
| **Bubble Sort** | Right â†’ Left | Largest unsorted element "bubbles" to its final position at the end | Look for: largest elements settling at the **right end** one by one (7, then 6, then 5...) |
| **Selection Sort** | Left â†’ Right | Finds minimum (or maximum) and places it in the next position | Look for: smallest elements being placed at the **left side** one by one (1, then 2, then 3...) |
| **Insertion Sort** | Left â†’ Right | Takes next element and inserts it into correct position within sorted portion | Look for: elements being **inserted anywhere** in the sorted portion (not just ends), building sorted array from left |
